```python
import pandas as pd
import json
```

```python
with open('/content/drive/MyDrive/CAPSTONE/generatif/intents.json', 'r') as f:
    data = json.load(f)
```

```python
respon = pd.DataFrame(data['intents'])
respon = respon.responses
respon
```

```python
dic = {"tag":[], "patterns":[], "responses":[]}
for example in data['intents']:
    for pattern in example['patterns']:
        dic['patterns'].append(pattern)
        dic['tag'].append(example['tag'])
        dic['responses'].append(example['responses'])
```

```python
pd.options.display.max_colwidth=15
df = pd.DataFrame.from_dict(dic)
df
```

```python
translated_data=[]
```

```python
for i in range(123,len(df)):
  if i!=29 :
    inten = []
    translate_tag = translator.translate(df['tag'][i],src='en',dest='id')
    translate_tag = translate_tag.text
    translate_pattern = translator.translate(df['patterns'][i],src='en',dest='id')
    translate_pattern = translate_pattern.text
    list_respon=[]
    for j in range(len(df['responses'][i])):
      translate_respon = translator.translate(df['responses'][i][j],src='en',dest='id')
      translate_respon = translate_respon.text
      list_respon.append(translate_respon)
    inten = [translate_tag,translate_pattern,list_respon]
    translated_data.append(inten)

```
